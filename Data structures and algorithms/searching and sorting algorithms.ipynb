# Linear search algorithm 
- This is a metod for finding an element in a data set (list/array) where each element is checked one by one until the required element is found.

### How it works
- Starting from  the beginning of the dataset,each element is examined  and compared to the target until a match is found.
- Once  tha match is found the search ends.
- The index of the target element is returned.
-if  no element is equal to the target element,we return -1 as the element isn't found.


### Pseudocode
- Procedure Linear Search(Array,Value)
    - FOR each item in Arrray
       - IF item == Value
        - return index of  item
       - END IF
    - END FOR
    - return -1 
- END PROCEDURE

### Time Complexity Analysis

Best Case :O(1)
- the element is  present at the first index. 
Worst Case:O(N) 
- the element might be present at the last index. where N is the size of the list.
Average Case: O(N)
- The element is present at any index

### Advantages
- Best for small datasets.
- doesn't require additional memory
- can be used whether the dataset is sorted or not.

### Disadvantages
- Not suitable for large datasets.
- slow since it has a time complexity of O(N) therefore the search time grows linearly with the size of the dataset.

### When to use it
- When the dataset is small.
- Seraching elements in a dataset stored in contiguous memory

### When not to use it
- When the datset is large,dealing with sorted data or frequent searches.







```python
### python implementation
def search(arr,x):
    for i in range (len(arr)):
        if arr[i] == x:
            return i
    return -1
arr = [1,2,3,4,5,10]
x= 5
result = search(arr,x)
if result == -1:
    print("Element is not present in array")
else:
    print("Element is present at index",result)


```

# Binary Search Algorithm/Logarithmic search/binary chop
- This is used to find the position of a target  in a sorted array by repeatedly dividing the interval in half.The search interval is halved by comparing the target element with the middle value of the search space.

### How it works.
- The search space is divided into two halves by finding the middle index.
- the middle index is compared to the target.
- If the middle index is the target the such ends.
- if the  target is smaller than the middle element ,search in the left half.
- If larger search in the right half.
- Until the search space is exhausted.



### Pseudocode

- function binary_search(array,target):
    - left = 0
    - right= length(array) -1 
    - WHILE left <= right:
        - mid = (left + right)  // 2
        - IF array[mid] == target:
            - return mid
        - ELIF array[mid] < target:
            - left = mid +1
        - ELSE:
            - right = mid-  1 
    return - 1             
      


### Time complexity 
Logarithmic O(logN)
- It works by dividing the search space in half repeatedly.
- And eliminating  half of the other elements after comparison between the target and mid element.


- Best Case: O(1)
The middle element is the target.

- Average Case: O(log N)
- Worst Case: O(log N)
Since the search space is halved.


### Advantages
- faster to use for large data sets
- More efficient to searching algorithms with the similar time complexities.(exponential search)
- Most suitable for searching data in hard drives and the cloud.

#### Disadvantages
- The Array should sorted
- The data structure being searched must be stored in contiguous memory(an array).
- The elements have to be comparable.e.g letters in alphabetical order.


### when to use it
- Searching in a sorted (list/array),large datasets,fastlookups(dictionaries)
- Search problems(guessing a number)

### when not to use it
- Dealing with small datasets,unsorted data,data that changes frequently



```python
### Binary search
# It can be implemented recursively/iteratively
#Iterative
def binary_search(array,target):
    left = 0
    right = len(array) - 1

    while (left <= right):
        mid = (right + left) // 2
        if array[mid] == target:
           return mid
        elif array[mid] < target : 
            left = mid + 1
        else:
            right = mid - 1 
     
array = [1, 3, 4, 10, ]
target = 10

result = binary_search(array, target)
if result == -1:
    print("Element is not present in array")
else:
    print("Element is present at index", result)



```

# Sorting algorithms
- Used to rearrange  data in either ascending/descending order.
### Bubble sort
- Repeatedly compares adjacent elements and swaps them if they are in the wrong order.
# How it works(using multiple passes)
- Two adjacent elements are compared
- if the first one is greater than the second one,we passit to the 
end( its correct position)
- We move onto the next pair and repeat the same action until the list is sorted.
- In each pass ,each element in a pair is compared .


### pseudocode.
- function bubbleSort(arr):
    - n = len(arr)
    -  for i in range(n):
        -  swapped = False
        - for j in range(0, n-i-1):
            - IF arr[j] > arr[j+1]:
                - arr[j], arr[j+1] = arr[j+1], arr[j]
                - swapped = True 
        - IF (swapped == False):
            - break   

### Time complexity                
- Quadratic O(n^2)
Involves nested loops that are repeatedly compare and swap elements

Worst-case time complexity: O(n²)
- when the array is sorted in reverse order.

Best-case time complexity: O(n) 
- when the array is already sorted, due to early exit.

Average-case time complexity: O(n²) 
- Similar to worst case.

Advantages 
-  easy to understand and implement.
- does not require any additional memory space.
- It is a stable sorting algorithm

Disadvantages of Bubble Sort:
- has a time complexity of O(n2) which makes it very slow for large data sets.
-  has almost no or limited real world applications.(used in academics to teach different ways of sorting.)

# When to use 
- Small datasets,nearly sorted data,where memory is extremely limited(since it works in place)

# When not to use
- Large datasets,systems that require fast responses.




```python
### Bubble sort implementation
def bubbleSort(arr):
    n = len(arr)
    for  i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True 
        if (swapped == False):
            break   

arr = [20,9,6,11]  
bubbleSort(arr)
print("Sorted array:",arr)      
```

# Inserion sort
- Works by iteratively inserting each element of an unsorted list into its corrected position in a sorted portion of the list.
- start from the second element(index 1)in the array.The first element is assumed to be sorted.
- Compare second element with the first element and check if the second element is smaller then swap them.
- Move to the third element and compare it with the first two elements and put at its correct position
- Repeat until the entire array is sorted.

### pseudocode
- function insertionSort(arr):
    - for i = 1 to length(arr)-1:
        - key = arr[i]
        - j = i-1
        - WHILE j> =0 and arr[j] > key:
            - arr[j + 1] = arr[j]
            - j= j-1
        - arr[j+1] = key    
- END FUNCTION


### Time complexities
Best case O(n):
- The array is already sorted

Worst Case O(n^2)
- The array is in reverse order.

Average Case O(n^2)
- A random order of elements

### Advantages
- Simple and easy to implement.
- Cannot easily be altered(stable).
- Suitable for small  and nearly sorted datasets.
- an in-place algorithm(no need for additional memory).
- Adoptive as the number of inversions is directly proportional to number of swaps. 

### Disadvantages 
- Inefficient for large datasets.
- Not as efficient as other sorting algorithms

### When to use it
- Small datasets,nearly sorted data,online sorting


### When  not to use it
- Large datasets,randomly ordered data
     


```python
# Implementation for insertion sort
def  insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j= j-1
        arr[j+1] = key   
    return  arr
arr = [4, 11, 13, 5, ]
sorted_arr = insertion_sort(arr)
print("Sorted array:", sorted_arr)
```

# Selection sort
- works by comparing elements to determine their order
Repeatedly selects the smallest/largest element from an unsorted portion of a list moving it to the sorted potion.
it works by first scanning the list to find the smallest number and swaps it with the first position.
- moves to the second position and repeats the same process.
- The sorted portion grows from the left as the unsorted portion shrinks.
- Until the array is sorted.

### Pseudocode
- PROCEDURE SelectionSort(arr)
    - FOR i ← 0 TO length(arr)-1
        - min ← i
        - FOR j ← i+1 TO length(arr)-1
            - IF arr[j] < arr[min]
                - min ← j
            - END IF
        - END FOR
        - SWAP arr[i] with arr[min]
    - END FOR
- END PROCEDURE



### Time complexity
- is O(n²) in all cases—best, average.
- always checks every element in the unsorted section, even if the list is already sorted

- One loop selects an element in an array= O(n)
- Another loop to compare that element with every other Array element = O(n)
- Overall time complexity is = O(n^2)


### Advantages
- Easy to understand and implement
- Has in place memory
- requires less number of swaps


### Disadvantages
- has a time complexity of O(n^2) making  it slower.
- Might end up swapping  equal elements (unstable)

### When to use it
- Small datasets,few swaps, datasets with O(1) memory space.

### When not to use it
- Large datasets,stable sorting,adaptive sorting




```python
### Implementation for selction sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr = [64, 25, 12, 22, 11]
sorted_arr = selection_sort(arr)
print("Sorted array:", sorted_arr)
```

# Quick sort

- Quicksort follows a divide-and-conquer strategy. It works by picking a "pivot" element from the array, partitioning the other elements into two groups—those less than the pivot and those greater than it—and then recursively sorting the two groups.


### How it works
- Select an element as a pivot
- Partition the array with smaller elements on the pivot's left and greater ones on its right
- apply the same process to left and right of the pivot.
- Recursion stops when there is only one element left.

### pseudocode
- function quickSort(arr)
    - IF length of arr <= 1
        - return arr
    - END iF 

    - pivot = arr[length of arr // 2]
    - left = []
    - middle = []
    - right = []

    - for each x in arr
        - IF x < pivot
            - append x to left
        - ELSE IF  x == pivot
            - append x to middle
        - ELSE
            - append x to right
        - END IF
    - END FOR

    - return concatenate(quickSort(left), middle, quickSort(right))
- END function




### Time complexity
Best Case: O(n log n)
-  occurs when the pivot always splits the array into two roughly equal halves
Average Case θ(n log n)
- the pivot divides the array into two parts, but not necessarily equal.
Worst Case: (O(n²))
- Occurs when the smallest or largest element is always chosen as the pivot .(unbalanced partitions)

### Advantages
- In place sorting
- Fast general purpose algorithm due its average time complexity.
- Cache friendly due to sequential accessing of data.
- Suitable for large datasets.
- Easier to solve problems due to divide and conquer

### Disavantages
-  unstable since equal elements swap positions and their relative order isn't maintained.
- not suitable for small datasets.
- Worst time complexity if pivot is poorly chosen. 



### When to use it
- Large datasets,in place ,general purpose sortng,cache memory

### When  not to use it
- Nearly sorted data,small datasets,stability requirement



```python
### quick sort implementataion
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

array = [3, 6, 8, 10, 1, 2, 1]
print("Sorted array:", quick_sort(array))
```

# Merge sort
divides the array into two halves, sort each half, and then merge the sorted halves back together. 
The  process is repeated until the entire array is sorted.

### How it works
- the list or array is divided into two halves.
-  Each subarray is sorted using the merge sort algorithm
- The sorted subarrays are merged back together in sorted order.

### Pseudocode
- PROCEDURE merge(arr, left, mid, right)
    - n1 ← mid - left + 1
    - n2 ← right - mid

    - L ← array of size n1
    - R ← array of size n2

    - FOR i FROM 0 TO n1-1
        - L[i] ← arr[left + i]
    - END FOR

    - FOR j FROM 0 TO n2-1
        - R[j] ← arr[mid + 1 + j]
    - END FOR

    - i ← 0
    - j ← 0
    - k ← left

    - WHILE i < n1 AND j < n2
        - IF L[i] <= R[j]
            - arr[k] ← L[i]
            - i ← i + 1
        - ELSE
            - arr[k] ← R[j]
            - j ← j + 1
        - END IF
        - k ← k + 1
    - END WHILE

    - WHILE i < n1
        - arr[k] ← L[i]
        - i ← i + 1
        - k ← k + 1
    - END WHILE

    WHILE j < n2
        arr[k] ← R[j]
        j ← j + 1
        k ← k + 1
    - END WHILE
- END PROCEDURE

- PROCEDURE merge_sort(arr, left, right)
    - IF left < right
       - mid ← (left + right) // 2

        - merge_sort(arr, left, mid)
        - merge_sort(arr, mid + 1, right)
        - merge(arr, left, mid, right)
    - END IF
- END PROCEDURE





### Time complexity
Best Case: O(n log n)
-  the array is already sorted or nearly sorted.

Average Case: O(n log n)
-  the array is randomly ordered.

Worst Case: O(n log n)
-  the array is sorted in reverse order.


### Advantages
- stable
- Performs well on large data sets due to  worst case time complexity 
- Easy to implement

### Disadvantages
- not in place(requirs additional memory tor store merged sub - arrays)
- Slow


### When to use
- stable sorting is required,large data sets,linked lists,external sorting.
### when not to use it
- Small data sets,in place,cache performance








```python
### Merge sort implementation
def merge_sort(arr):

    if len(arr) <= 1:

        return arr

    

    mid = len(arr) // 2

    left_half = arr[:mid]

    right_half = arr[mid:]

    

    left_half = merge_sort(left_half)

    right_half = merge_sort(right_half)

    

    return merge(left_half, right_half)



def merge(left, right):

    result = []

    i = 0

    j = 0

    while i < len(left) and j < len(right):

        if left[i] < right[j]:

            result.append(left[i])

            i += 1

        else:

            result.append(right[j])

            j += 1

    result += left[i:]

    result += right[j:]

    return result



# Example usage

arr = [3, 2, 1, 4, 5, 6, 7]

sorted_arr = merge_sort(arr)

print(sorted_arr)
```

```python

import random
import time
import matplotlib.pyplot as plt

# Sorting Algorithms
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Wrapper to time sorting functions
def time_sorting_algorithm(algorithm, arr):
    start_time = time.time()
    if algorithm == quick_sort:
        sorted_arr = algorithm(arr)  # Handle Quick Sort separately
    else:
        algorithm(arr)
    end_time = time.time()
    return end_time - start_time

# Generate random datasets of varying sizes
def generate_random_data(size):
    return random.sample(range(size * 10), size)  # Random numbers without repetition

# Automate testing of sorting algorithms on large datasets
def automate_sorting_tests(data_sizes, algorithms):
    results = {}
    for algorithm_name, algorithm in algorithms:
        times = []
        for size in data_sizes:
            data = generate_random_data(size)
            times.append(time_sorting_algorithm(algorithm, data.copy()))
        results[algorithm_name] = times
    return results

# Plot the results
def plot_results(results, data_sizes):
    plt.figure(figsize=(10, 6))
    for algorithm_name, times in results.items():
        plt.plot(data_sizes, times, label=algorithm_name)
    plt.xlabel("Dataset Size")
    plt.ylabel("Time (seconds)")
    plt.title("Sorting Algorithm Performance on Large Datasets")
    plt.legend()
    plt.grid(True)
    plt.show()

# Main function to execute the entire automated process
def main():
    data_sizes = [10000, 50000, 100000, 200000, 500000]
    algorithms = [
        ("Bubble Sort", bubble_sort),
        ("Selection Sort", selection_sort),
        ("Insertion Sort", insertion_sort),
        ("Merge Sort", merge_sort),
        ("Quick Sort", quick_sort),
    ]
    results = automate_sorting_tests(data_sizes, algorithms)
    for algorithm_name, times in results.items():
        print(f"{algorithm_name}: {times}")
    plot_results(results, data_sizes)

if __name__ == "__main__":
    main()
```
