### Algorithm analysis
- This is the process of evaluating the performance of an algorithm based its time and space complexity. It is simply evaluating an algorithm's performance by measuring how much time and memory it requires to execute  based on the size of its input.


Asymptotic analysis  is the study of change in performance of algorithm with the change in the order of the input size.


#Time complexity
- This represents how many operations an algorithm can perform in realtion to the input size.It is how long an algorithm takes to run as the size of input increases.It is expressed using Big O notation.

#Space complexity
- This is the measure of amount of memory used by an algorithm to solve a problem or execute an algorithm in realtion to input size.
A low space complexity means that an algorithm requires relatively little memory to solve the problem while a high space complexity means that an algorithm requires a lot of memory to solve the problem which can lead to a slow performance.It is also expressed using Big O notation.


#Correctness,Termination and effectiveness
- Correctness means that an algorithm will produce a desired output for a valid input.
- Termination means that an algorithm will have to stop executing and not run into an infinite loop.
- Effectiveness is how well an algorithm solves a problem in terms of correctness,efficiency and scalability.An algorithm should [produce correct output while using minimal resources.]


#Algorithm complexity/Running time/complexity and measurement
- Algorithm complexity is the measure of how long and how much  an algorithm would take to solve a problem given  an  input of size n.
- It is measured in terms of time and space complexity  asymptotically using big O notation.(bigO,bigalpha and big theta).
Time complexity is measured by counting the number of basic operations and analyzing how they grow with n.Space complexity measures additional memory used beyond input storage.Measurement is the comparison  and analyzation of the efficiency of different algorithms.










### catergories of algorithm complexity
Constant time(O(1))
- An algorithm takes a fuxed amount of time to execute regardless of the input size.

Logarithmic time(O(log n))
- An algorithm executes a problem proportional to the algorithm of the input size n.

Linear time(O(n))
- The execution time grows directly proportional to the input size.

Quadratic time(O(n**2))
- The execution time grows proportional to the square of the input size.e.g nested loops

Exponential time(O(2**n))
- The execution time grows exponentially with the input size.
(Considered as inefficient)

Factorial time(O(n!))
- The execution time grows as the factorial of the input size input size.(Considered as inefficient)

Quadratic time(O(n**3))
- The execution time grows proportional to the cube of the input size.

Summary of Common Complexities
- O(1): Accessing an element in an array.
- O(log n): Binary search.
- O(n): Iterating through a list to find the maximum element.
- O(n log n): Merge sort or quicksort.
- O(n²): Bubble sort or selection sort.
- O(2^n): Recursive solutions to problems like the Fibonacci sequence.
- O(n!): Brute-force permutations.

- Dependent quadratic loops

### Big-O notation(Order-of)
- Big-O notation is a mathematical notation that describes the upper bound of the running time(time complexity ) of an algorithm.It analyzes and gives  the worst case situation or scenario  of an algorithm's performance.It provides an upper limit on the time taken by an algorithm in terms of size of the input.It is denoted by O(f(n)) where f(n) is a function that represents the number of steps an algrithm performs to solve a problem of size n.

Importance of Big O notation
- It helps analyze the efficiency of algorithms
- Allows programmers to compare different algorithms and chise the most efficient one for a specific problem.
- It enables description of how runtime or space requirements of an algorithm grow as the input increases.
- It helps in understanding the scability of algorithms and predicts how they will perform as input size grows.


How to determine big O notation
- Big O notation is a mathematical notation used to describe the asymptotic behavior of a function as its input grows infinitely large. .

Steps to Determine Big O Notation:

- Identify the Dominant Term:
Examine the function and identify the term with the highest order of growth as the input size increases.
Ignore any constant factors or lower-order terms.

- Determine the Order of Growth:
The order of growth of the dominant term determines the Big O notation.


- Write the Big O Notation:
The Big O notation is written as O(f(n)), where f(n) represents the dominant term.
For example, if the dominant term is n^2, the Big O notation would be O(n^2).


- Simplify the Notation (Optional):
In some cases, the Big O notation can be simplified by removing constant factors or by using a more concise notation.
For instance, O(2n) can be simplified to O(n).


- Always remove constant factors.


Example:
Consider the following function: f(n) = 3n^2 + 2n + 8
- Identify the Dominant Term: The dominant term is 3n^2, as it grows
- Determine the Order of Growth: The order of growth of 3n^2 is n^2.
- Write the Big O Notation: The Big O notation is O(n^2).
- Simplify the Notation (Optional): The Big O notation can be simplified to O(n^2).







### Algorithm runtime efficiency
Algorithm runtime efficiency is all about how effectively an algorithm performs in terms of time and space.
- Best Case: The scenario where the algorithm performs the fastest.

- Worst Case: The scenario where the algorithm performs the slowest.

- Average Case: The expected performance over all possible inputs.



### Comparisons of different complexity functions

- constant Time	O(1)-Best performance unaffected by input size	e.g Accessing a single element in an array

- Logarithmic Time	O(log n)	-Very efficient, grows slowly e.gBinary search in a sorted array

- Linear Time	O(n)	- Efficient, grows linearly with input size e.g	Iterating through an array

- Linearithmic Time O(n log n)	- Moderate, grows faster than linear but still reasonable	e.gEfficient sorting algorithms like Merge Sort and Quick Sort

- Quadratic Time	O(n^2)	Poor performance, grows quadratically	e.g Simple sorting algorithms like Bubble Sort and Insertion Sort

- Cubic Time	O(n^3)	Very poor, grows cubically e.g 	Algorithms with three nested loops

- Exponential Time	O(2^n)	Terrible, grows exponentiallye.g	Solving the Traveling Salesman Problem using brute force

- Factorial Time	O(n!)	Extremely terrible, grows factorially e.g	Permutations of a set of elements
