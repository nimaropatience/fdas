### 2.Abstract Data Types and Data Structures
Abstract data types are the specifications of data types.It focuses on what the data type does and not how it does.

The key roles  of ADT include :
- Abstraction and Simplification :This focuses on what the data type does and not how it does it and also shields users from internal workings from the data type.It also provides a simplified interface for interacting with data.

- Encapsulation: ADT'S store data and operations done on data into a single unit and ensure data protection and integrity.

- Data structure independence: Implementations by the data structure on an ADT can be done without affecting the code and adjustments on specific needs can be made without extensive code modifications.

- Modularity and Reusability: Complex systems can be built from smaller,independent units and it easiy to understand and reuse code in different parts of a program.

- Information hiding: It does not allow access to internal data ad only exposes an interface  to a user and protects data from malicious changes.

- Software Engineering principles: it encourages good softwre engineering practces by promoting abstraction, modularity and code reusability.

### The differences between ADT and data structures are as follows:

- A Data structure deals with the representation of data in memory while an ADT deals with the behavior and operations of a data type.

- An ADT is independent of any specific implementation while a Data structure is a concrete realization of an ADT.

- Examples of ADT'S inclued lists,stacks, queues and graphs while examples of data structures include arrays,linked lists,hash tables and binary tables.

- An ADT describes what a data type can do and the operations which can be performed on it  while a data structure implements a data type by specifying how the data is organized and stored in memory along with the algorithms for manipulating it.

- ADT is a  specification  while a Data structure is an implementation of that specification.

### examples of ADTS and how they are implemented

A list ,stack and queue can be implemented using linked lists and arrays.
A dictionary  can be implemented using a hash table and a balanced search tree.




### 3.Types of Data structures 
-  Arrays
Arrays are data structures that store a fixed-size sequence of elements of the same type. 

Each element can be accessed using an index. Arrays are used in programming to store and manage collections of data. 

- Linked lists
Linked lists are a type of data structure that consists of a sequence of elements, called nodes.
each node contains data and  a link to the next node in the sequence.

- Stacks are a type of data structure that follow the Last In, First Out (LIFO) principle. 
This means that the most recently added element is the first one to be removed.

append: Adding an element to the top of the stack.

Pop: Removing the top element from the stack.

Peek/Top: Accessing the top element without removing it.

IsEmpty: Checking if the stack is empty.

- A queue is a data structure that follows the First In, First Out (FIFO) principle.
This means that the first element added to the queue will be the first one to be removed.

- Recursion is a process in which a function calls itself directly or indirectly to solve a problem. 

- A tree is a hierarchical data structure consisting of nodes, where each node has a value and references to its child node.


- Graphs are powerful data structures used to model relationships between entities. 
graphs can be implemented using dictionaries, adjacency lists, or adjacency matrices.















```python

#Arrays
#Example
# Creating an array of integers
numbers = [1, 2, 3, 4]

# Accessing elements in the array
print(numbers[0])  

# Modifying an element in the array
numbers[2] = 10
print(numbers) 


#linked lists

#Example
linked_list = []
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
print("linked_list",linked_list)

#Stacks


#Example of a Stack:
stack =[]
stack.append(1)
stack.append(2)
stack.append(3)
print("stack",stack) 
print(stack.pop()) 

#examples of queues,recursion and trees
#Queues

queue =[]
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
print("queue",queue)
if queue:
    print(queue.pop(0))


#Recursion 
# An example of recursion is calculating the factorial of a number.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Example usage
print(factorial(5))  

#Trees

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child):
        self.children.append(child)

    def __repr__(self):
        return f"Node({self.value})"

# Create nodes
root = Node("A")
node_b = Node("B")
node_c = Node("C")
node_d = Node("D")
node_e = Node("E")

# Build tree
root.add_child(node_b)
root.add_child(node_c)
node_b.add_child(node_d)
node_b.add_child(node_e)

# Example of accessing the tree structure
print(root.value)  # Root node
for child in root.children:
    print(child.value)  # Children of root

for child in node_b.children:
    print(child.value)  # Children of node B


# GRAPHS

# Example usage
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def print_graph(self):
        for node in self.graph:
            print(f"{node} -> {self.graph[node]}")

# Example usage
g = Graph()
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(3, 3)
g.print_graph()




```

### 4.Operations in data structures
- Traversing (or traversal) refers to the process of visiting each element or node in a data structure, such as an array, linked list, tree, or graph, exactly once to perform some operation.

- Searching  refers to the process of finding a specific element within a data structure.

The goal is to determine whether the element exists and, if so, to locate its position .  

- Insertion in data structures refers to the operation of adding an element to a data structure.

- Deletion  refers to the operation of removing an element from a data structure

The specifics of the deletion operation depend on the type of data structure being used.

- Updating  refers to the operation of modifying an existing element within the data structure.

- Sorting is the process of arranging elements in a particular order, usually in ascending or descending order

- Merging is the process of combining two or more sorted lists or arrays into a single sorted list


```python
### Examples
# Transversing
array = [1, 2, 3, 4, 5]
for element in array:
    print(element)

#searching
def linear_search(arr, target):
    for index, value in enumerate(arr):
        if value == target:
            return index
    return -1

# Example usage
arr = [4, 2, 7, 1, 9]
target = 7
print(linear_search(arr, target))  # Output: 2

#Insertion 
arr = [1, 2, 4, 5]
index = 2
element = 3
arr.insert(index, element)
print(arr)  # Output: [1, 2, 3, 4, 5]

#deletion

arr = [1, 2, 3, 4, 5]
index = 2
del arr[index]
print(arr)  # Output: [1, 2, 4, 5]

#Updating

arr = [1, 2, 3, 4, 5]
index = 2
new_value = 10
arr[index] = new_value
print(arr)  # Output: [1, 2, 10, 4, 5]

#Sorting

# Example usage of sorted()
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = sorted(arr)
print("Sorted array:", sorted_arr)  # Output: [11, 12, 22, 25, 34, 64, 90]

#merging

def merge_sorted_lists(list1, list2):
    merged_list = []
    i = j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append remaining elements of list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Append remaining elements of list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
result = merge_sorted_lists(list1, list2)
print("Merged List:", result)  # Output: [1, 2, 3, 4, 5, 6, 7, 8]

```

# SHORT NOTES ON ALGORITHMS

- Brute force

It tries all kinds of possible combinations to find a solution to a problem

- Recursive

A problem is broken down into subproblems and keeps calling itself again and again using the same base.


- Divide and conquer

A problem is  divided into two sections.One section breaks down the problem into subproblems.Section two solves the subproblems independently.
The combination of these two sections is the problem solved.(Done by section two)


- Dynamic programming algorithm

It stores previously calculated results to avoid calculating them over and over.


- Greedy algorithm

The solution to the problem is built part by part.
The decision to chose the next part is based on the immediate benefit it gives.


- Backtracking algorithm

The problem is solved incrementally one piece at a time and removes those solutions that fail to satisfy the constraints of the problem at any point.



### 6.Algorithm representation
- Pseudocode is a simplified, informal way of describing an algorithm or a program's logic using plain language and basic programming constructs. It's a bridge between human thinking and actual code. Pseudocode does not follow the syntax rules of any specific programming language, but it provides a clear and concise representation of the steps needed to solve a problem or perform a task.

### how pseudocode is used to represent algorithms
- Pseudocode breaks down an algorithm into individual steps and describes them in human language.
- Pseudocode uses simple constructs like loops, conditionals, and assignments that are common to most programming languages, making it easy to translate into actual code.
- Pseudocode organizes the logic of the algorithm in a structured way, making it easy to follow and understand.


- A flowchart is a diagrammatic / graphical representation of an algorithm




### 7.Guidelines for drawing  flowcharts
- Use the standard flowchart symbols consistently to ensure that the flowchart is universally understandable.

-  Label each symbol clearly with concise and descriptive text.

-  Maintain a consistent direction of flow (typically top-to-bottom or left-to-right) to make the flowchart easier to follow
.
-  Keep the flowchart simple and avoid clutter. Break complex processes into smaller, manageable sub-processes.

-  Ensure that the flow of the process is logical and easy to follow. Use arrows to guide the reader through the process.

-  Minimize crossed lines to avoid confusion. Use connectors or off-page references if necessary.

-  Use colors and different line styles sparingly to highlight important sections or differentiate between different types of actions.

- Include a title and a legend (if necessary) to provide context and explain any non-standard symbols or conventions used
